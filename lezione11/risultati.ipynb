## Esercizio 1

Lo scopo di questo esercizio è quello di capire come i parametri $N_{epochs}$, $N_{train}$ e $\sigma$, che rappresentano rispettivamente il numero di epoche, il numero di dati di training e il rumore utilizzati per allenare una rete neurale ad attuare una regressione lineare, influenzano l'abilità di tale rete neurale ad attuare tale regressione lineare.

La seguente cella produce $N_{train}$ numeri distribuiti uniformemente tra -1 e 1 che corrispondono alle ascisse dei dati di training e $N_{train}/10$ numeri che corrispondono alle ascisse dei dati di test. Successivamente, produce le ordinate $y$ per ciascuna $x$ utilizzando la formula $y=mx+b$ con $m=2$ e $b=1$ con rumore $\sigma$. Poi, crea la rete neurale costituita da uno strato denso con una singola unità e la compila utilizzando come ottimizzatore la discesa stocastica del gradiente e come funzione costo e metrica il mean squared error. Infine, svolge il training della rete neurale con i dati di training utilizzando una batch size pari a 32 e un numero di epoche pari a $N_{epochs}$ e stampa i parametri $m$ e $b$ stimati dal modello.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects
# parametri f(x) = m*x + b
m = 2 # coefficiente angolare
b = 1 # intercetta

#rete neurale e training
np.random.seed(0)
x_train = np.random.uniform(-1, 1, 1000)
x_valid = np.random.uniform(-1, 1, 100)
x_valid.sort()
y_target = m * x_valid + b # valori ideali di y
sigma = 0 # rumore
y_train = np.random.normal(m * x_train + b, sigma)
y_valid = np.random.normal(m * x_valid + b, sigma)
# creazione rete neurale
model = tf.keras.Sequential()
model.add(Dense(1, input_shape=(1,)))
# compilazione modello
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
# fit modello con dati
# 30 epoche e batch size 32
history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=30,
          shuffle=True, # shuffle dei dati
          validation_data=(x_valid, y_valid))
data = model.get_weights()
print(data)
```

Dopo aver svolto il training della rete neurale, è possibile usare la rete neurale per stimare le ordinate relative a delle ascisse. La seguente cella disegna in blu la retta $y=2x+1$ e in rosso i punti appartenenti a tale retta previsti dalla rete neurale. Naturalmente, la regressione lineare risulta essere migliore se il parametro $m$ stimato dalla rete neurale risulta essere vicino a 2 e $b$ a 1.

```python
x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)
plt.show()
```

Utilizzando le due celle precedenti, è stato osservato il loro output per capire come varia il loro risultato a seconda di $N_{epochs}$, $N_{train}$ e $\sigma$. Le osservazioni ottenute sono le seguenti:

- all'aumentare di $N_{epochs}$, la stima dei parametri $m$ e $b$ della rete neurale migliora e si ottiene una regressione lineare migliore: infatti, mantenendo fissati $\sigma=0$ e $N_{train}=500$, si ottiene che, per $N_{epochs}=20$, $m=1.8379083$ e $b=0.9963284$, per $N_{epochs}=35$, $m=1.93979$ e $b=0.9992806$, per $N_{epochs}=50$, $m=1.9936666$ e $b=0.9999222$, per $N_{epochs}=200$, $m=1.9999942$ e $b=0.9999999$. All'aumentare di $N_{epochs}$ la stima di $m$ si avvicina sempre di più a 2 e quella di $b$ a 1. Ciò è spiegato dal fatto che all'aumentare del numero di epoche, aumenta il training svolto dalla rete neurale e pertanto si ottiene una regressione lineare migliore;

- all'aumentare di $N_{train}$, la stima dei parametri $m$ e $b$ della rete neurale migliora e si ottiene una regressione lineare migliore: infatti, mantenendo costanti $\sigma=0$ e $N_{epochs}=20$, si ottiene che, per $N_{train}=500$, $m=1.8223153$ e $b=0.99677074$, per $N_{train}=1000$, $m=1.9723978$ e $b=0.999661$, per $N_{train}=2000$, $m=1.9997146$ e $b=1.0000035$, per $N_{train}=5000$, $m=1.9999943$ e $b=1.0000001$. La regressione lineare migliora all'aumentare di $N_{train}$ poiché la rete neurale si allena con una quantità maggiore di dati e pertanto riesce a fittare meglio i dati;

- per quanto riguarda il rumore, la qualità della regressione lineare, fissando $N_{epochs}=30$ e $N_{train}=500$, dipende da quanto vale $\sigma$: infatti, per $\sigma=0.1$, $m=1.899089$ e $b=0.9877944$, per $\sigma=0.3$, $m=1.9260144$ e $b=0.96584654$, per $\sigma=0.5$, $m=2.013977$ e $b=0.9446528$, per $\sigma=1$, $m=1.9721148$ e $b=0.8937194$. Come si può notare dai dati riportati, all'aumentare del rumore la stima dell'intercetta peggiora, mentre la stima del coefficiente angolare migliora per $\sigma$ sufficientemente bassi, tuttavia peggiora se $\sigma$ diventa troppo grande.

Con queste osservazioni, è terminato lo studio della dipendenza della qualità della regressione lineare da $N_{epochs}$, $N_{train}$ e $\sigma$. Riassumendo, per effettuare un buon fit, è necessario scegliere un numero sufficientemente alto di epoche e di dati di training e un rumore non troppo forte.

## Esercizio 2

Lo scopo di questo esercizio è creare un modello per ottenere un fit della funzione $f(x)=4-3x-2x^2+3x^3$ nell'intervallo $[-1,1]$ e osservare come tale modello fitta dati al di fuori di tale intervallo.

La seguente cella inizialmente crea i dati utilizzati per il training e il test della rete neurale con un rumore gaussiano pari a $\sigma=0.1$. Il numero di dati utilizzati per il training è $N_{train}=10000$, mentre per il test 1000.  Successivamente, la cella crea una rete neurale sequenziale strutturata nel seguente modo: ha tre strati iniziali densi composti da 50 unità con funzione di attivazione relu, poi ha un ultimo strato denso che restituisce l'output. La cella compila il modello utilizzando come ottimizzatore la discesa stocastica del gradiente e come funzione costo e metrica il mean squared error. Infine, la cella svolge il training della rete neurale: per il training sono state impostate 150 epoche e un batch size pari a 32. Le informazioni relative al training sono state salvate nella variabile history in modo da poter disegnare successivamente il grafico del costo in funzione del numero di epoche.

```python
# parametri funzione
a = 4
b = -3
c = -2
d = 3

#rete neurale e training
x_train = np.random.uniform(-1, 1, 10000)
x_valid = np.random.uniform(-1, 1, 1000)
x_valid.sort()
y_target = a + b*x_valid + c*(x_valid**2) + d*(x_valid**3) # valori ideali di y
sigma = 0.1 # rumore
y_train = np.random.normal(a + b*x_train + c*(x_train**2) + d*(x_train**3), sigma)
y_valid = np.random.normal(a + b*x_valid + c*(x_valid**2) + d*(x_valid**3), sigma)
# creazione rete neurale
model2 = tf.keras.Sequential()
model2.add(Dense(50, activation='relu', input_shape=(1,)))
model2.add(Dense(50, activation='relu'))
model2.add(Dense(50, activation='relu'))
model2.add(Dense(1))
# compilazione modello
model2.compile(optimizer='sgd', loss='mse', metrics=['mse'])
# fit modello con dati
# 300 epoche e batch size 32
history = model2.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=150,
          shuffle=True, # shuffle dei dati
          validation_data=(x_valid, y_valid))
```

Di seguito è riportato il grafico del costo in funzione del numero di epoche sia per i dati di training sia per i dati test.

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

Il grafico in precedenza mostra come la performance del modello migliora a seconda del numero di epoche utilizzate: infatti, minore è il costo e maggiore è la qualità del fit effettuato. Per l'ultima epoca è stato ottenuto che il costo per i dati di training vale 0.0104, mentre quello per i dati test 0.0109. La performance ottenuta con i dati di training è simile a quella ottenuta con i dati test, quindi la rete neurale ha una buona capacità di fittare dati che non ha mai visto.

Per visualizzare graficamente il funzionamento della rete neurale, nella seguente cella è disegnato un grafico: la linea blu rappresenta la funzione esatta $f(x)=4-3x-2x^2+3x^3$ nell'intervallo $[-1,1]$ e i punti rossi rappresentano i punti previsti dalla rete neurale.



```python
x_predicted = np.random.uniform(-1, 1, 200)
y_predicted = model2.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)
plt.show()
```

Come si può notare dal risultato della cella precedente, la rete neurale riesce a fittare i dati con la funzione $f(x)=4-3x-2x^2+3x^3$ nell'intervallo $[-1,1]$.

La cella di seguito disegna in blu la funzione esatta $f(x)=4-3x-2x^2+3x^3$ nell'intervallo $[-3,-1]$ e in rosso i punti previsti dalla rete neurale in tale intervallo per verificare la performance del modello al di fuori dell'intervallo dove è stato allenato.

```python
x_predicted2 = np.random.uniform(-3, -1, 200)
y_predicted2 = model2.predict(x_predicted2)
x_valid2 = np.random.uniform(-3, -1, 100)
x_valid2.sort()
y_target2 = a + b*x_valid2 + c*(x_valid2**2) + d*(x_valid2**3)
plt.scatter(x_predicted2, y_predicted2,color='r')
plt.plot(x_valid2, y_target2)
plt.grid(True)
plt.show()
```

Come si può notare dal risultato della cella, la rete neurale è incapace di eseguire il fit della funzione $f(x)=4-3x-2x^2+3x^3$ al di fuori dell'intervallo dei dati su cui è stata allenata. Naturalmente, non essendo stata allenata su tale intervallo, la rete neurale non può eseguire un buon fit: affinché sia in grado di effettuare tale fit, è necessario estendere i dati di training all'intervallo $[-3,-1]$.

Dallo svolgimento di questo esercizio, si è giunti alle seguenti conclusioni:

- per creare una buona rete neurale che effettui il fit di una funzione, è necessario scegliere un numero di parametri né troppo basso né troppo alto: nel primo caso si ottiene un costo alto e quindi un pessimo fit, nel secondo caso si rischia di overfittare i dati di training e quindi il modello non predice bene i dati che non ha mai visto;

- dopo aver scelto il numero di parametri, bisogna scegliere una batch size e un numero di epoche adatto per far in modo che il costo relativo ai dati di training e test sia sufficientemente basso e simile: infatti, se il costo relativo tra i dati di training e test è basso e la loro differenza bassa, il modello ha una capacità buona di predire dati che non ha mai visto;

- bisogna evitare di usare la rete neurale per effettuare fit di dati al di fuori dell'intervallo su cui è stata allenata in quanto non è in grado di predire la disposizione di punti al di fuori dell'intervallo di training.

## Esercizio 3

L'esercizio 3 richiede di fare il fit della funzione $f(x,y)=\sin(x^2+y^2)$ nell'intervallo $x \in [-3/2,3/2]$ e $y \in [-3/2,3/2]$.

La seguente cella genera 10000 dati di training e 1000 dati test con un rumore gaussiano $\sigma = 0.1$ per allenare e testare la rete neurale. Successivamente, crea una rete neurale con la stessa struttura di quella dell'esercizio 2 e la compila usando come ottimizzatore la discesa stocastica del gradiente e come funzione costo e metrica il mean squared error. Infine, la cella esegue il training della rete neurale: per il training è stato scelto una batch size pari a 32 e un numero di epoche pari a 150.

```python
#rete neurale e training
x_train = np.random.uniform(-1.5, 1.5, 10000)
x_valid = np.random.uniform(-1.5, 1.5, 1000)
y_train = np.random.uniform(-1.5, 1.5, 10000)
y_valid = np.random.uniform(-1.5, 1.5, 1000)
x_valid.sort()
y_valid.sort()
X_train = np.column_stack((x_train, y_train))
X_valid = np.column_stack((x_valid, y_valid))
sigma = 0.1 # rumore
z_train = np.random.normal(np.sin(x_train**2+y_train**2), sigma)
z_valid = np.random.normal(np.sin(x_valid**2+y_valid**2), sigma)
# creazione rete neurale
model3 = tf.keras.Sequential()
model3.add(Dense(50, activation='relu', input_shape=(2,)))
model3.add(Dense(50, activation='relu'))
model3.add(Dense(50, activation='relu'))
model3.add(Dense(1))
# compilazione modello
model3.compile(optimizer='sgd', loss='mse', metrics=['mse'])
# 300 epoche e batch size 32
history = model3.fit(X_train, z_train, 
          batch_size=32, epochs=150,
          shuffle=True, # shuffle dei dati
          validation_data=(X_valid, z_valid))
```

Di seguito è disegnato il grafico del costo in funzione del numero di epoche sia per i dati di training sia per i dati test.

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

Da come si può notare dai risultati delle due celle precedenti, il costo per i dati di training e quello per i dati test sono simili per l'ultima epoca: infatti, per $N_{epochs}=150$ si ha che il primo vale 0.0099 e il secondo 0.0108. Ciò significa che il modello ha una buona capacità di predire dati che non ha mai visto.

Per visualizzare graficamente il corretto funzionamento della rete neurale, la seguente cella disegna in blu la funzione esatta $f(x,y)=\sin(x^2+y^2)$ nell'intervallo $x \in [-3/2,3/2]$ e $y \in [-3/2,3/2]$ e in rosso i punti previsti dalla rete neurale sempre nello stesso intervallo.

```python
# Definizione della funzione
def f(x, y):
    return np.sin(x ** 2 + y ** 2)

# Generazione dei dati
x_predicted = np.random.uniform(-1.5, 1.5, 800)
y_predicted = np.random.uniform(-1.5, 1.5, 800)
X_predicted = np.column_stack((x_predicted, y_predicted))
z_predicted = model3.predict(X_predicted)
x = np.linspace(-1.5, 1.5, 200)
y = np.linspace(-1.5, 1.5, 200)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
# Creazione del grafico 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
ax.scatter3D(x_predicted, y_predicted, z_predicted, color='red')
# Etichette degli assi
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(x,y)')

# Visualizzazione del grafico
plt.show()
```

Come si può notare dal risultato della cella, i punti previsti dalla rete neurale si dispongono correttamente sul grafico di $f(x,y)$, quindi la rete neurale funziona correttamente.
