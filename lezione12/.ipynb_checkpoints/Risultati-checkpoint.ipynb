## Esercizio 1

```python
import tensorflow as tf
from tensorflow import keras
import os
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import SGD, Adam, RMSprop, Adagrad, Adadelta, Adam, Adamax, Nadam
#impostazione seme
seed=0
np.random.seed(seed)
tf.random.set_seed(seed)
img_rows, img_cols = 28, 28 #numero pixel
num_classes = 10 #numero di possibili output
(X_train, Y_train), (X_test, Y_test) = mnist.load_data() #caricamento dati
X_train = X_train.reshape(X_train.shape[0], img_rows*img_cols) #trasformo in vettori di dim 28x28
X_test = X_test.reshape(X_test.shape[0], img_rows*img_cols)
X_train = X_train.astype('float32') #converto in float
X_test = X_test.astype('float32')
X_train /= 255 #riscalo dati su intervallo [0,1]
X_test /= 255
Y_train = keras.utils.to_categorical(Y_train, num_classes) #converto vettori in matrici binarie
Y_test = keras.utils.to_categorical(Y_test, num_classes)

def create_DNN():
    # inizializzo modello
    model = Sequential()
    # aggiunta di un layer
    model.add(Dense(400,input_shape=(img_rows*img_cols,), activation='tanh'))
    # aggiunta di un layer
    model.add(Dense(100, activation='elu'))
    # aggiunta di un layer
    model.add(Dense(100,activation='tanh'))
    # aggiunta dropout 0.5
    model.add(Dropout(0.5))
    # layer finale
    model.add(Dense(num_classes, activation='softmax')) #devono uscire uno dei 10 numeri
    
    return model

def compile_model():
    # creazione modello
    model=create_DNN()
    # compilazione modello
    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer=SGD(),
                  metrics=['acc'])
    return model
```

```python
# allenamento del modello
batch_size = 32 # dimensione batch
epochs = 50 # numero di epoche

# creazione rete neurale
model_DNN = compile_model()

# allenamento DNN e salva le informazioni di training in history
history = model_DNN.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(X_test, Y_test))
```

```python
# valutazione del modello
score = model_DNN.evaluate(X_test, Y_test, verbose=1)

# stampa performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# grafico accuracy
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.ylabel('model accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='best')
plt.show()

# grafico loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.ylabel('model loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='best')
plt.show()
```

```python
# verifica predizione
X_test = X_test.reshape(X_test.shape[0], img_rows*img_cols)
predictions = model_DNN.predict(X_test)

X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols,1)

plt.figure(figsize=(15, 15)) 
for i in range(30):    
    ax = plt.subplot(2, 30, i + 1)    
    plt.imshow(X_test[i, :, :, 0], cmap='gray')    
    plt.title("D:{}\nP:{}".format(np.argmax(Y_test[i]), np.argmax(predictions[i])))     
    plt.axis('off') 
plt.show()
# D rappresenta il label della cifra disegnata e P la cifra prevista dalla rete neurale
```

## Esercizio 2

```python
from keras.layers import Flatten, Conv2D, MaxPooling2D
# trasformo i dati
if keras.backend.image_data_format() == 'channels_first':
    X_train = X_train.reshape(X_train.shape[0], 1, img_rows, img_cols) #un solo canale perché non ho colori
    X_test = X_test.reshape(X_test.shape[0], 1, img_rows, img_cols)
    input_shape = (1, img_rows, img_cols)
else:
    X_train = X_train.reshape(X_train.shape[0], img_rows, img_cols, 1)
    X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols, 1)
    input_shape = (img_rows, img_cols, 1)

# creazione rete neurale convoluzionale
def create_CNN():
    # inizializzazione modello
    model = Sequential()
    # primo layer convoluzionale con 10 filtri
    model.add(Conv2D(10, kernel_size=(5, 5),
                     activation='relu',
                     input_shape=input_shape))
    # aggiunta layer di pooling
    model.add(MaxPooling2D(pool_size=(5, 5)))
    # Aggiungi il layer Flatten
    model.add(Flatten())
    # aggiunta di un layer
    model.add(Dense(100,activation='tanh'))
    # aggiunta dropout 0.5
    model.add(Dropout(0.5))
    # layer finale
    model.add(Dense(num_classes, activation='softmax')) #devono uscire uno dei 10 numeri
    
    # compilazione modello
    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer='SGD',
                  metrics=['acc'])
    return model
```

```python
# allenamento del modello
batch_size = 32 # dimensione batch
epochs = 50 # numero di epoche

# creazione rete neurale
model_CNN = create_CNN()

# allenamento CNN e salva le informazioni di training in history2
history2 = model_CNN.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(X_test, Y_test))
```

```python
# valutazione del modello
score = model_CNN.evaluate(X_test, Y_test, verbose=1)

# stampa performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# grafico accuracy
plt.plot(history2.history['acc'])
plt.plot(history2.history['val_acc'])
plt.ylabel('model accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='best')
plt.show()

# grafico loss
plt.plot(history2.history['loss'])
plt.plot(history2.history['val_loss'])
plt.ylabel('model loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='best')
plt.show()
```

```python
# verifica predizione
X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols, 1)
predictions2 = model_CNN.predict(X_test)

X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols,1)

plt.figure(figsize=(15, 15)) 
for i in range(30):    
    ax = plt.subplot(2, 30, i + 1)    
    plt.imshow(X_test[i, :, :, 0], cmap='gray')    
    plt.title("D:{}\nP:{}".format(np.argmax(Y_test[i]), np.argmax(predictions2[i])))     
    plt.axis('off') 
plt.show()
# D rappresenta il label della cifra disegnata e P la cifra prevista dalla rete neurale
```
